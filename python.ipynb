```python
# converting suscribers , video views and population, urban population in millions; 
df['Subscribers'] = (df['Subscribers'] / 1000000).round(2)
df['Video Views'] = (df['Video Views'] / 1000000).round(2)
df["Population"] = (df['Population'] / 1000000).round(2)
df['Urban Population'] = (df['Urban Population']/1000000).round(2)
df.tail()
```

```python
# finding the duplicates values;
df.duplicated().sum()
```

```python
# finding number of unique rows;
df[-df.duplicated()].shape[0]
```

```python
# handling missing values in dataset
NaNnumerical= df.select_dtypes(include=["float64","int64"] ).columns
NaNcategorical= df.select_dtypes(include=["object"]).columns
df[NaNnumerical]= df[NaNnumerical].fillna(0)
df[NaNcategorical]= df[NaNcategorical].fillna("Unknown")
```

# Data analysis and visualization;


```python
#Insight 1: T-Series leads with the highest YouTube subscribers in India,
# followed by entertainment and music channels like SET India and Zee Music Company.
#Root Cause:Dominance of music and entertainment content suggests a strong preference for
#these genres among Indian viewers.
#Provable Solution: Content creators targeting the Indian audience should consider producing music and
# entertainment content to tap into the popularity of these genres and potentially gain more subscribers.

# find the top 10 indian youtube subscribers;
indian_channels = df[df['Abbreviation'] == 'IN']
top_10_indian_channels = indian_channels.sort_values(by='Subscribers', ascending=False).head(10)

print(top_10_indian_channels[['Youtuber', 'Subscribers']])

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(top_10_indian_channels['Youtuber'], top_10_indian_channels['Subscribers'], color='blue')
plt.xlabel('YouTube Channels')
plt.ylabel('Subscribers (in millions)')
plt.title('Top 10 Indian YouTube Channels by Subscribers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# category wise subscribers in percentage in horizontal bar chart;
#Insight 2:Music and Entertainment categories have the highest percentage of subscribers, 
#indicating a strong audience preference for these genres.
#Root Cause:The popularity of music and entertainment content may be due to their universal appeal and 
#ability to engage a wide audience.
#Provable Solution:Content creators should strategically incorporate music and entertainment elements into their videos
#aligning with the preferences of the majority of subscribers, to maximize engagement and subscriber growth.

import matplotlib.pyplot as plt
import seaborn as sns

category_subscribers = df.groupby('Category')['Subscribers'].sum()
total_subscribers = df['Subscribers'].sum()

percentage_subscribers = (category_subscribers / total_subscribers) * 100
percentage_subscribers = percentage_subscribers.reset_index()

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Subscribers', y='Category', data=percentage_subscribers, color='skyblue')

# Add data labels
for index, value in enumerate(percentage_subscribers['Subscribers']):
    bar_plot.text(value, index, f'{value:.2f}%', va='center')

plt.xlabel('Percentage of Subscribers')
plt.ylabel('Category')
plt.title('Category-wise Distribution of Subscribers')

# Horizontal bar chart with data labels
plt.show()
```

```python
# Insight 3:The distribution of country-wise subscribers reveals that the United States 
# has the highest percentage of subscribers at 43.9%, followed by India at 25.8%, Unknown regions at 17.5%, 
# Brazil at 7.3%, and the United Kingdom at 5.5%.
# Root Cause: Variations in subscriber percentages can be attributed to factors such as population size, 
# internet penetration, cultural interests, and the popularity of YouTube content in each country.
# Provable Solution:Localized Content Develop Create content that aligns with the cultural preferences and 
# interests of the dominant market 
# Pie chart of country-wise subscribers
country_subscribers = df.groupby('Country')['Subscribers'].sum().sort_values(ascending=False).head(5)
plt.pie(country_subscribers, labels=country_subscribers.index, autopct='%1.1f%%',
        colors=['gold', 'lightskyblue', 'lightcoral', 'lightgreen', 'lightpink'])

plt.title('Country-wise Subscribers Distribution')
plt.show()
```

```python
#Insight 4:KIMPRO tops the list with the highest average yearly earnings, followed by Boom and T-Series,
# indicating varied revenue streams and successful monetization strategies
#Root Cause:Diverse revenue sources, including sponsorships, merchandise, and affiliate marketing,
#contribute to the high average earnings.
#Provable Solution:Content creators should explore multiple revenue streams beyond ad revenue, 
#such as collaborations, merchandise sales, and affiliate partnerships, to optimize their overall yearly earnings.
# Top 10 youtubers by average yearly earnings:
import matplotlib.pyplot as plt


temp = df[['Youtuber', 'AverageYearlyEarnings']].sort_values(by='AverageYearlyEarnings', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(temp['Youtuber'], temp['AverageYearlyEarnings'], color='red')
plt.xlabel('Youtuber')
plt.ylabel('Average Yearly Earnings')
plt.title('Top 10 Youtubers by Average Yearly Earnings')
plt.xticks(rotation=45, ha='right')

# Data labels
for index, value in enumerate(temp['AverageYearlyEarnings']):
    plt.text(index, value, f"${value:.2f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
#Insights 16: Comparison between India and US with Entertainment and Music category popularity:

# Dominance of Music in India: The top 5 YouTubers in the Music category have a significant share of subscribers in India and US.
# Music content resonates strongly with the Indian audience.

# Root Cause:Cultural Preference for Music India has a rich cultural heritage with a strong emphasis on music and entertainment.
# Factors: Cultural influence, diverse music genres, and widespread music appreciation.

# Provable Solution: Strategy: Content creators targeting the Indian audience should align with the cultural preference for music.
# Action: Explore collaborations with musicians, cover popular songs, and produce content that integrates cultural elements to maximize engagement and subscriber growth.


# Both Entertainment and Music categories have diverse genres represented in the top 5 YouTubers in the US.
# US audiences have varied interests across different content genres.

# Root cause: Hollywood influence, celebrity culture, and the prominence of entertainment content.

# Provable Solution: Collaborate with influencers, leverage popular trends, and create engaging and shareable content to 
# maintain and grow the subscriber base.Content creators targeting the US audience should embrace diversity and personalize 
# content.



# Filter rows for Entertainment and Music categories in India and US
filtered_df = df[df['Category'].isin(['Entertainment', 'Music']) & df['Abbreviation'].isin(['IN', 'US'])]

# Separate the data for each country and category
india_entertainment = filtered_df[(filtered_df['Abbreviation'] == 'IN') & (filtered_df['Category'] == 'Entertainment')]
india_music = filtered_df[(filtered_df['Abbreviation'] == 'IN') & (filtered_df['Category'] == 'Music')]
us_entertainment = filtered_df[(filtered_df['Abbreviation'] == 'US') & (filtered_df['Category'] == 'Entertainment')]
us_music = filtered_df[(filtered_df['Abbreviation'] == 'US') & (filtered_df['Category'] == 'Music')]

# Define a function to get the top 5 YouTubers
def get_top_5(dataframe):
    return dataframe.nlargest(5, 'Subscribers')[['Youtuber', 'Subscribers', 'Video Views']]

# Get the top 5 YouTubers for each category and country
top_5_india_entertainment = get_top_5(india_entertainment)
top_5_india_music = get_top_5(india_music)
top_5_us_entertainment = get_top_5(us_entertainment)
top_5_us_music = get_top_5(us_music)

# create figure:
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# pie charts each category and country
def plot_pie(ax, labels, sizes, title):
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    ax.axis('equal') 
    ax.set_title(title)

plot_pie(axes[0, 0], top_5_india_entertainment['Youtuber'], top_5_india_entertainment['Subscribers'], 'Top 5 Entertainment YouTubers (India)')
plot_pie(axes[0, 1], top_5_india_music['Youtuber'], top_5_india_music['Subscribers'], 'Top 5 Music YouTubers (India)')
plot_pie(axes[1, 0], top_5_us_entertainment['Youtuber'], top_5_us_entertainment['Subscribers'], 'Top 5 Entertainment YouTubers (US)')
plot_pie(axes[1, 1], top_5_us_music['Youtuber'], top_5_us_music['Subscribers'], 'Top 5 Music YouTubers (US)')


plt.tight_layout()
plt.show()
```
